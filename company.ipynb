```python
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# isin()
# find the most destructive death over batsman in the history of IPL
# Strike Rate = (Number of runs/Number of balls)/100
# min batsman 200 balss in over 16-20

```

```python
death_over=delivery[delivery['over']>15]
```

```python
death_over.shape
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
```

```python
x=all_batsman>200
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
#runs scored by all these 43 batsman
#balls played by all these 43 batsman
final=death_over[death_over['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```

```python
sr.sort_values(ascending=False).head(10)
```
